```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

random_values = np.random.rand(5)

total_random = sum(random_values)
normalized_probabilities = [value / total_random for value in random_values]


cumulative_probabilities = np.cumsum(normalized_probabilities)


probabilities_df = pd.DataFrame({
    'Alan': [f"Alan {i+1}" for i in range(5)],
    'Rastgele Değer': random_values,
    'Normalized Probability': normalized_probabilities,
    'Cumulative Probability': cumulative_probabilities
})

print("Probabilities Tablosu:")
display(probabilities_df)


def roulette_wheel_selection(cumulative_probabilities):
    
    pick = np.random.rand()
    for i, cumulative_prob in enumerate(cumulative_probabilities):
        if pick < cumulative_prob:
            return i, pick  

selected_index, selection_value = roulette_wheel_selection(cumulative_probabilities)
print(f"Seçilen Alan: Alan {selected_index + 1}")
print(f"Seçim Değeri : {selection_value}")


elements = [f"Alan {i+1}" for i in range(5)]
plt.figure(figsize=(8, 8))
plt.pie(normalized_probabilities, labels=elements, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('5 Alan İçin Normalize Edilmiş Olasılıklar')
plt.show()
```
